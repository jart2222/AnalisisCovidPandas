```python
import pandas as pd
import numpy as np
```

## Extraer la información del archivo.

```python
count_vacci= pd.read_csv("./country_vaccinations.csv")
```

##  Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones puedes realizar con cada una de ellas, asegurándote que las columnas con fechas sean del tipo datetime64

```python
count_vacci['date'] = pd.to_datetime(count_vacci['date'])
count_vacci['total_vaccinations'] = count_vacci['total_vaccinations'].fillna(0)
count_vacci['people_vaccinated'] = count_vacci['people_vaccinated'].fillna(0)
count_vacci['people_fully_vaccinated'] = count_vacci['people_fully_vaccinated'].fillna(0)
count_vacci['daily_vaccinations_raw'] = count_vacci['daily_vaccinations_raw'].fillna(0)
count_vacci['daily_vaccinations'] = count_vacci['daily_vaccinations'].fillna(0)
count_vacci['total_vaccinations_per_hundred'] = count_vacci['total_vaccinations_per_hundred'].fillna(0)
count_vacci['people_vaccinated_per_hundred'] = count_vacci['people_vaccinated_per_hundred'].fillna(0)
count_vacci['people_fully_vaccinated_per_hundred'] = count_vacci['people_fully_vaccinated_per_hundred'].fillna(0)
count_vacci['daily_vaccinations_per_million'] = count_vacci['daily_vaccinations_per_million'].fillna(0)
count_vacci.info()
```

## Determinar la cantidad de vacunas aplicadas de cada compañía (con base en cómo lo reporta cada país en la columna vaccines, en otras palabras, agrupe por vaccines y realice la sumatoria)

```python
count_vacci[['vaccines', 'total_vaccinations']].groupby('vaccines').sum()
```

## Obtener la cantidad de vacunas aplicadas en todo el mundo

```python
count_vacci[['total_vaccinations']].sum()
```

## Calcular el promedio de vacunas aplicadas por país

```python
count_vacci[['country', 'total_vaccinations']].groupby('country').mean()
```

## Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.

```python
count_vacci.loc[count_vacci['date']=='2021-01-29'][['total_vaccinations']].sum()
```

## Crear un dataframe nuevo denominado `conDiferencias`que contenga los datos originales y una columna derivada (`diferencias`) con las diferencias de aplicación entre las columnas `daily_vaccinations` y `daily_vaccinations_raw`.

```python
conDiferencia=count_vacci.copy()
conDiferencia['diferencias']= np.abs(conDiferencia['daily_vaccinations']-conDiferencia['daily_vaccinations_raw'])
```

```python
conDiferencia.head()
```

## Obtener el periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua.

```python
print("Max_date: ",count_vacci["date"].max(), " Min: ", count_vacci["date"].min())
```

## Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter , ).

```python
conCantidad=count_vacci.copy()
conCantidad['canVac']=count_vacci['vaccines'].str.split(', ').str.len()
```

```python
conCantidad.head()
```

## Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.

```python
antes20=count_vacci.loc[count_vacci['date']<'2020-12-20']
antes20.head()
```

## Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfize

```python
pfizer= count_vacci.query('vaccines.str.contains("Pfizer")', engine='python')
pfizer.head()
```

## Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente.

```python
with pd.ExcelWriter("resultadosReto.xlsx") as writer:
    conDiferencia.to_excel(writer, sheet_name="conDiferencia")
    conCantidad.to_excel(writer, sheet_name="conCantidad")
    antes20.to_excel(writer, sheet_name="antes20")
    pfizer.to_excel(writer, sheet_name="pfizer")
```
